```python
import torch
import matplotlib.pyplot as plt
```

```python
data = torch.load('./checkpoints/skinnet/checkpoint.pth.tar')
statsrec = data["state_dict"]
plt.plot(statsrec[0], 'r', label = 'training loss', )
plt.plot(statsrec[1], 'b', label = 'test loss' )
plt.legend(loc='lower right')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('Training and test loss')
```
